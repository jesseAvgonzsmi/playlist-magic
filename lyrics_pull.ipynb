```python
import time
from requests.exceptions import Timeout

def get_lyrics(song, artist, retries=5):
    for i in range(retries):
        try:
            response = genius.search_song(song, artist)
            if response is None or response.lyrics == "Specified song does not contain lyrics. Rejecting.":
                print(f"No lyrics found for {song} by {artist}")
                return "No Lyrics"
            return response.lyrics
        except Timeout:
            if i < retries - 1: 
                time.sleep(1)
                continue
            else:
                raise
```

```python
lyrics = pd.read_csv('lyrics.csv')
```

```python
import re

def extract_lyrics(text):
    start = text.find("\n") + len("\n")
    return text[start:].strip()

def clean_lyrics(text):
    if text.count(" - ") > 5:
        return None  # Drop the text
    text = re.sub(r'\[.*?\]', '', text)  # Remove anything inside []
    text = re.sub(r'\d+', '', text)  # Remove numbers
    text = text.replace("Embed","")
    return text.replace("\n", " ")

lyrics['lyrics'] = lyrics['lyrics'].apply(extract_lyrics)
lyrics['lyrics'] = lyrics['lyrics'].apply(clean_lyrics)
lyrics.head(5)
```

```python
lyrics.to_csv('lyrics.csv', index=False)
```
