## Features and Tracks

```python
import numpy as np
chunk_size = 45
df['chunk'] = np.arange(len(df)) // chunk_size

# chunk_size = 95
# df['chunk'] = np.arange(len(df)) // chunk_size
```

```python
def chunk_id_list(x):
    return df[df['chunk'] == x]['track.id'].tolist()
```

```python
def get_tracks(ids):
    try:
        response = sp.tracks(ids)
        return response
    except SpotifyException as e:
        if e.http_status == 429:
            # print(f"Rate limit exceeded, sleeping for {e.retry_after} seconds.")
            # time.sleep(e.retry_after)
            raise Exception("Rate limit exceeded. Stopping the program.")
        else:
            raise

def get_audio_features(ids):
    try:
        response = sp.audio_features(ids)
        return response
    except SpotifyException as e:
        if e.http_status == 429:
            print(f"Rate limit exceeded, sleeping for {e.retry_after} seconds.")
            time.sleep(e.retry_after)
            raise Exception("Rate limit exceeded. Stopping the program.")
        else:
            raise

# # Usage
# for index, row in df.iterrows():
#     id = row['track.id']
#     try:
#         response = get_tracks(id)
#         # process response...
#     except Exception as e:
#         print(str(e))
#         break
```

# Artists

```python
tracks = pd.read_csv('tracks.csv')
```

```python
#from the artist column, pull out the artist id
tracks['artist.id'] = tracks['artists'].str.split(',').str[2]
#from artist.id remove 'id': and the quotes
tracks['artist.id'] = tracks['artist.id'].str.replace("'", '')
#replace id: with nothing
tracks['artist.id'] = tracks['artist.id'].str.replace('id:', '')
#strip the white space from the artist.id column
tracks['artist.id'] = tracks['artist.id'].str.strip()

tracks.to_csv('tracks.csv', index=False)
```

```python
artists_unique = tracks['artist.id'].unique().tolist()
artists_unique
```

```python
#create a new dataframe from the unique artist ids
artists_df = pd.DataFrame(artists_unique, columns=['artist.id'])
artists_df
```

```python
#break artists_dfs into chunks of 50
import numpy as np
chunk_size = 50
artists_df['chunk'] = np.arange(len(artists_df)) // chunk_size
artists_df.head(10)
```

```python
def get_artists(ids):
    try:
        response = sp.artists(ids)
        return response
    except SpotifyException as e:
        if e.http_status == 429:
            print(f"Rate limit exceeded, sleeping for {e.retry_after} seconds.")
            time.sleep(e.retry_after)
            raise Exception("Rate limit exceeded. Stopping the program.")
        else:
            raise
```

```python
def chunk_id_list(x):
    return artists_df[artists_df['chunk'] == x]['artist.id'].tolist()
```

```python
ids = chunk_id_list(0)
response = get_artists(ids)
response2 = json_normalize(response['artists'])
response2
```

```python
artist_df = pd.DataFrame()

for i in range(0, 32):
    print(i)
    ids = chunk_id_list(i)
    artists = get_artists(ids)
    artists2 = json_normalize(artists['artists'])
    artist_df = pd.concat([artist_df, artists2])
    time.sleep(15)

artist_df
```
